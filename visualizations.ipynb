```python
import numpy as np
%matplotlib inline

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import ImageGrid
from matplotlib import cm

from common import *
```

```python
X, y = load_detpep10exp_dataset(20, sample_range=(0, 1))
```

## 3D plot

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
im = ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap=cm.jet)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('x3')
plt.colorbar(im, fraction=0.026, pad=0.04)
plt.show()
```

### X1-X2 plane

```python
fig = plt.figure(figsize=(15, 15))
axs = ImageGrid(fig, 111,
                nrows_ncols = (1, 3),
                axes_pad = 0.5,
                cbar_location = "right",
                cbar_mode="single",
                cbar_size="5%",
                cbar_pad=0.05)
x1 = np.linspace(0, 1, 21)
x2 = np.linspace(0, 1, 21)
xx1, xx2 = np.meshgrid(x1, x2)
for i, x3 in enumerate([0.1, 0.5, 0.9]):
    y = detpep10exp(xx1, xx2, x3)
    im = axs[i].imshow(y)
    im.set_clim(vmin=0, vmax=40)
    axs[i].set_xlabel('x1')
    axs[i].set_ylabel('x2')
    axs[i].set_xticks([0, 10, 20], labels=[0, 0.5, 1])
    axs[i].set_yticks([0, 10, 20], labels=[0, 0.5, 1])
    axs[i].set_title('x3={}'.format(x3))
plt.colorbar(im, cax=axs.cbar_axes[0])
plt.show()
```
